# Clustering 

```python
import numpy as np
import pandas as pd
```

```python
mnist_data = pd.read_csv(r"G:\Data sets\Clustering\mnist_data.csv")

mnist_data.head()
```

```python
mnist_data.shape
```

```python
im = np.asarray(mnist_data.iloc[0:1,:]).reshape(28,28)

import matplotlib.pyplot as plt

plt.imshow(im,cmap = plt.cm.plasma)
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters = 10)
model.fit(mnist_data)
```

```python
print(np.unique(model.labels_))
```

```python
# Exploring the cluster 1

cluster1 = mnist_data[model.labels_ == 0]

cluster1_imgs = cluster1.iloc[[np.random.randint(0,cluster1.shape[0]) for i in range(0,5)]]

for i in range(0,cluster1_imgs.shape[0]):
    plt.subplot(1,5,i+1)
    img_fig = np.asarray(cluster1_imgs[i:i+1]).reshape(28,28)
    plt.imshow(img_fig,cmap = plt.cm.plasma)
```

```python
# Exploring the cluster 2

cluster2  = mnist_data[model.labels_==1]

cluster2_img = cluster2.iloc[[np.random.randint(0,cluster2.shape[0]) for i  in range(0,5)]]

for i in range(0,cluster2_img.shape[0]):
    plt.subplot(1,5,i+1)
    img_fig = np.asarray(cluster2_img[i:i+1]).reshape(28,28)
    plt.imshow(img_fig,cmap = plt.cm.Reds)
```

```python
# Exploring cluster 8

cluster8 = mnist_data[model.labels_ == 7]
cluster8_img = cluster8.iloc[[np.random.randint(0,cluster8.shape[0]) for i in range(0,5)]]

for i in range(0,cluster8_img.shape[0]):
    plt.subplot(1,5,i+1)
    img_fig = np.asarray(cluster8_img[i:i+1]).reshape(28,28)
    plt.imshow(img_fig,cmap = plt.cm.magma)
```
